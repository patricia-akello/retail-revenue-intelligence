# RETAIL REVENUE INTELLIGENCE & PROFITABILITY OPTIMIZATION
# PHASE 2: DATA CLEANING & FEATURE ENGINEERING

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
pd.set_option("display.float_format", "{:,.2f}".format)
```

## 2. Load Dataset

```python
df = pd.read_csv("Superstore.csv", encoding="latin1")

print("Dataset loaded successfully.")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

df.head()
```

## 3. Standardize Column Names

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

print("Standardized columns:")
print(df.columns.tolist())
```

## 4. Initial Data Inspection

```python
print("Dataset Shape:")
print(df.shape)

print("\nData Types:")
print(df.dtypes)

print("\nDataset Info:")
df.info()

print("\nSummary Statistics:")
display(df.describe(include="all"))
```

## 5. Missing Value Analysis

```python
missing_summary = pd.DataFrame({
    "missing_count": df.isnull().sum(),
    "missing_percentage": (df.isnull().sum() / len(df)) * 100
}).sort_values(by="missing_count", ascending=False)

display(missing_summary)
```

## 6. Duplicate Analysis

```python
total_duplicates = df.duplicated().sum()
print(f"Total duplicate rows: {total_duplicates}")

if total_duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicate rows removed.")
else:
    print("No duplicate rows found.")
```

## 7. Data Type Correction

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["ship_date"] = pd.to_datetime(df["ship_date"], errors="coerce")

numeric_columns = ["sales", "quantity", "discount", "profit"]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

print("Corrected data types:")
print(df.dtypes)
```

## 8. Date Quality Validation

```python
invalid_order_dates = df["order_date"].isnull().sum()
invalid_ship_dates = df["ship_date"].isnull().sum()

print(f"Invalid order dates: {invalid_order_dates}")
print(f"Invalid ship dates: {invalid_ship_dates}")

df["shipping_days"] = (df["ship_date"] - df["order_date"]).dt.days

invalid_shipping_days = df[df["shipping_days"] < 0]

print(f"Rows with negative shipping days: {len(invalid_shipping_days)}")
```

## 9. Business Logic Validation

```python
print("Negative sales records:", len(df[df["sales"] < 0]))
print("Negative quantity records:", len(df[df["quantity"] < 0]))
print("Discount below 0:", len(df[df["discount"] < 0]))
print("Discount above 1:", len(df[df["discount"] > 1]))

print("Loss-making transactions:", len(df[df["profit"] < 0]))
print("Zero-profit transactions:", len(df[df["profit"] == 0]))
print("Profitable transactions:", len(df[df["profit"] > 0]))
```

## 10. Remove Invalid Business Records If Any

```python
df = df[df["sales"] >= 0]
df = df[df["quantity"] >= 0]
df = df[(df["discount"] >= 0) & (df["discount"] <= 1)]
df = df[df["shipping_days"] >= 0]

print("Dataset shape after business validation:")
print(df.shape)
```

## 11. Feature Engineering: Date Features

```python
df["order_year"] = df["order_date"].dt.year
df["order_month"] = df["order_date"].dt.month
df["order_month_name"] = df["order_date"].dt.month_name()
df["order_quarter"] = df["order_date"].dt.quarter
df["order_year_month"] = df["order_date"].dt.to_period("M").astype(str)
df["order_day_of_week"] = df["order_date"].dt.day_name()
```

## 12. Feature Engineering: Profitability Metrics

```python
df["profit_margin"] = np.where(
    df["sales"] != 0,
    df["profit"] / df["sales"],
    0
)

df["profit_margin_percentage"] = df["profit_margin"] * 100

df["loss_making_flag"] = np.where(df["profit"] < 0, 1, 0)

df["profitability_status"] = np.select(
    [
        df["profit"] < 0,
        df["profit"] == 0,
        df["profit"] > 0
    ],
    [
        "Loss Making",
        "Break Even",
        "Profitable"
    ],
    default="Unknown"
)
```

## 13. Feature Engineering: Discount Segmentation

```python
df["discount_band"] = pd.cut(
    df["discount"],
    bins=[-0.01, 0, 0.10, 0.20, 0.40, 1.00],
    labels=[
        "No Discount",
        "Low Discount",
        "Moderate Discount",
        "High Discount",
        "Extreme Discount"
    ]
)
```

## 14. Feature Engineering: Revenue Segmentation

```python
df["revenue_category"] = pd.qcut(
    df["sales"],
    q=4,
    labels=[
        "Low Revenue",
        "Medium Revenue",
        "High Revenue",
        "Premium Revenue"
    ]
)
```

## 15. Feature Engineering: Shipping Performance

```python
df["shipping_speed"] = np.select(
    [
        df["shipping_days"] <= 2,
        df["shipping_days"].between(3, 5),
        df["shipping_days"] > 5
    ],
    [
        "Fast",
        "Standard",
        "Delayed"
    ],
    default="Unknown"
)
```

## 16. Feature Engineering: Customer Profitability

```python
customer_profitability = (
    df.groupby("customer_id")
    .agg(
        customer_total_sales=("sales", "sum"),
        customer_total_profit=("profit", "sum"),
        customer_order_count=("order_id", "nunique")
    )
    .reset_index()
)

customer_profitability["customer_profit_margin"] = np.where(
    customer_profitability["customer_total_sales"] != 0,
    customer_profitability["customer_total_profit"] / customer_profitability["customer_total_sales"],
    0
)

customer_profitability["customer_segment_risk"] = np.select(
    [
        customer_profitability["customer_total_sales"] >= customer_profitability["customer_total_sales"].quantile(0.75),
        customer_profitability["customer_total_profit"] < 0,
        customer_profitability["customer_profit_margin"] < 0.05
    ],
    [
        "High Value Customer",
        "Loss Making Customer",
        "Low Margin Customer"
    ],
    default="Standard Customer"
)

df = df.merge(
    customer_profitability[
        [
            "customer_id",
            "customer_total_sales",
            "customer_total_profit",
            "customer_order_count",
            "customer_profit_margin",
            "customer_segment_risk"
        ]
    ],
    on="customer_id",
    how="left"
)
```

## 17. Outlier Analysis

```python
def detect_outliers_iqr(dataframe, column):
    q1 = dataframe[column].quantile(0.25)
    q3 = dataframe[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = dataframe[
        (dataframe[column] < lower_bound) |
        (dataframe[column] > upper_bound)
    ]

    return lower_bound, upper_bound, len(outliers)

outlier_columns = ["sales", "quantity", "discount", "profit", "shipping_days"]

outlier_summary = []

for col in outlier_columns:
    lower, upper, count = detect_outliers_iqr(df, col)
    outlier_summary.append({
        "column": col,
        "lower_bound": lower,
        "upper_bound": upper,
        "outlier_count": count,
        "outlier_percentage": (count / len(df)) * 100
    })

outlier_summary_df = pd.DataFrame(outlier_summary)
display(outlier_summary_df)
```

## 19. Final Quality Check

```python
final_quality_check = pd.DataFrame({
    "missing_count": df.isnull().sum(),
    "missing_percentage": (df.isnull().sum() / len(df)) * 100,
    "data_type": df.dtypes
})

display(final_quality_check)

print("Final dataset shape:")
print(df.shape)

print("Final columns:")
print(df.columns.tolist())
```

## 20. Export Cleaned Dataset

```python
df.to_csv("superstore_cleaned.csv", index=False)

print("Cleaned dataset exported successfully as superstore_cleaned.csv")
```
